```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# creating graph variable
g = nx.Graph()
```

```python
# creating edge list
edges = [('Andre','Beverly'),
         ('Andre','Carol'),
         ('Andre','Diane'),
         ('Andre','Fernando'),
         ('Beverly','Andre'),
         ('Beverly','Garth'),
         ('Beverly','Diane'),
         ('Carol','Diane'),
         ('Carol','Fernando'),
         ('Diane','Fernando'),
         ('Diane','Garth'),
         ('Diane','Ed'),
         ('Ed','Garth'),
         ('Fernando','Garth'),
         ('Fernando','Heather'),
         ('Garth','Heather'),
         ('Heather','Ike'),
         ('Ike','Jane'),
        ]
```

```python
# adding the edge list to graph
g.add_edges_from(edges)
```

```python
# print graph
nx.draw(g, with_labels=True)
```
